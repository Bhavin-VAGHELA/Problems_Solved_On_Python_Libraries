# Bascis To Advance Practice problems on Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,30,40]
c=["magenta","yellow","green","blue"]
plt.bar(x,y,width=0.5,color=c,align="center",edgecolor="black",linewidth=2,alpha=1,label='bargraph')
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
